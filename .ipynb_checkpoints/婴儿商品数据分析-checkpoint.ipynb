# 婴儿商品数据分析  
## 一、 数据介绍  
数据来源：https://tianchi.aliyun.com/dataset/dataDetail?dataId=45 ，该数据集由天猫、淘宝提供。  
**字段含义：**  
1. (sample)sam_tianchi_mum_baby_trade_history.csv（购买商品历史记录——共7个字段，29971条数据）  
（1） user_id:用户id  
（2） auction_id:购买行为编号  
（3） cat_id:商品种类ID  
（4） cat1:商品属于哪个类别  
（5） property:商品属性  
（6） buy_mount:购买数量  
（7） day:购买时间  
2. (sample)sam_tianchi_mum_baby.csv（婴儿信息——共3个字段，953条数据）  
（1） user_id:用户id  
（2） birthday:出生日期  
（3） gender:性别（0 男性；1 女性）

## 二、 提出问题  
1. 婴儿的性别、年龄对用户购买有什么关联？对不同品类的商品有什么购买偏向？  
2. 热销商品的购买人群特征？  
3. 有没有季节依赖性较强的商品？有没有问题商品？

## 三、 数据清洗  
### 1. 列名重命名  
将列名重命名为中文格式，易于理解。 
### 2. 整合数据：  
通过Vlookup函数连接两张表，将商品交易表和婴儿信息表根据用户ID合并到一张表里，删除用不到的auction_id、property两列。  
### 3. 重复值处理  
通过【条件格式】【重复值】查找并删除重复id，本数据集并无重复id。  
### 4. 缺失值处理  
通过【查找和选择】【定位条件】查找缺失值，并删除缺失值，检查发现该数据集无缺失值。  
### 5. 一致化处理  
将各个字段数据类型更换成合适的类型，如将用户ID、商品种类数据类型换成文本类型，将购买日期、出生日期转换成日期格式；通过条件列将性别分为男、女、未知；计算出婴儿年龄，未出生的记为“未出生”，没有出生日期的记为“未知”。  
### 6. 异常值处理  
（1） 对购买数量进行排序，发现有一个用户的购买数量为1000，严重大于其他值，去除此异常值。  


（2） 按用户ID合并两个表，合并后有效数据只有954条。

（3） 合并后的数据中有8条数据缺失产品种类和购买次数，无效，因此选择去除。  
（4） 利用婴儿的出生日期和购买时间计算出购买商品时婴儿的年龄。   
计算购买时婴儿年龄发现了明显超出婴儿年龄范围的用户和大量在出生日期之前购买的用户，这部分用户无法判定是孕期购买还是出生日期信息错误，在此选择剔除这部分数据；因此选择剔除购买商品时婴儿年龄(天数)-365以下的值。此外，通过对婴儿年龄进行筛选，可以发现有一个最大年龄是28岁，显然，这是错误的，故将该行数据删除。

共940条记录。

## 四、 数据分析  
1. 商品类别与购买数量  
  
商品一共有6个分类，其中分类"28""50008168"的销量最佳。  
2. 人体结构关联  
（1） 性别  
  

男婴购买比例 = 931/488 = 1.91  
女婴购买比例 = 559/438 = 1.28  
总体来说男婴比女婴购买的多，细分的话主要体现在商品种类"28""50008168"上。  
（2） 年龄  
  
  
宝宝年龄越小需求就越大，特别是商品种类为"500014815"表现最为明显，使用商品"28""50008168"的婴儿年龄较大一点。

三、时间进度对销量影响

1.总体情况如下：  
  
可以发现2013年比2012年销量是有比较大的增加，主要是因为同年第二、三季度的销量有了巨大的提升。而2015年的第一季度同比减少约60%。  
2.将类别拆分，来寻找是否存在季节性商品  
  
  
结果发现商品类别为"28""50025520"的商品在第三第四季度的销量较大，以此推测这两类商品主要是用在这两个季度上。

## 五、结论分析

1. 婴儿的性别、年龄对用户购买有什么关联？对不同品类的商品有什么购买偏向？  
结论：男宝宝比女宝宝购买的更多，“50014815”更多购买的是男宝宝，“50008168“男女都比较喜欢；年龄越小的孩子对母婴商品需求量越大，“50014815”是0-1岁宝宝最喜欢的，“50008168“和”28“使用年龄更长。  
2. 热销商品的购买人群特征？  
结论：卖的最好的品类是“28“，但缺少大量用户信息；  
根据现有用户信息来看50014815”和“50008168“反倒更受欢迎，这两个都是0-1岁婴儿较喜欢的。  
3. 有没有季节依赖性较强的商品？有没有问题商品？  
结论：存在；商品类别为"28""50025520"的商品在第三第四季度销量比较多。
